```python
import pandas as pd
import numpy as np
```

```python
weather_dataset = pd.read_csv('wwo_normalized_manila\\worldweatheronline_normalized_manila_2017.csv', index_col=['datetime'], parse_dates=['datetime'], skipinitialspace=True)
# weather_dataset = weather_dataset[:35040]
weather_dataset
```

# Resampling
Resample the 15-minute interval dataset to 1-hour interval dataset.

```python
#convert to float
new_dataset = weather_dataset.astype(float)

new_dataset = new_dataset.resample('1H').interpolate(method='linear', axis=0).ffill().bfill()
new_dataset.to_csv('wwo_normalized_manila\\worldweatheronline_normalized_manila_2017_1hour.csv')
new_dataset
```

## Trying to resample daily rainfall
Resample the 15-minute interval dataset to 1-day interval dataset. Get average daily weather variable. 

```python
#convert to float
new_dataset = weather_dataset.astype(float)

# new_dataset = new_dataset.resample('1D').interpolate(method='linear', axis=0).ffill().bfill()
new_dataset = new_dataset.groupby(new_dataset.index.strftime("%Y-%m-%d")).mean()
new_dataset.to_csv('wwo_normalized_manila\\worldweatheronline_normalized_manila_2017_1day.csv')
new_dataset
```
